```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
#set all columns to be displayed
pd.set_option('display.max_columns', None)
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
X = df.drop("koi_disposition", axis=1)
y = df["koi_disposition"]
print(X.shape, y.shape)
```

```python
df.koi_disposition.unique()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
X_train.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
len(X_train_scaled[0])
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
```

```python
label_encoder.fit(y_test)
encoded_y_test = label_encoder.transform(y_test)
```

```python
 from tensorflow.keras.utils import to_categorical
```

```python
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
y_train_categorical
```

# Train the Model



```python
from sklearn.svm import SVC 
```

```python
# Fit to the training data and validate with the test data
model = SVC(C=1.0, cache_size=200, class_weight=None, coef0=0.0,
  decision_function_shape='ovo', degree=3, gamma='auto', kernel='linear',
  max_iter=-1, probability=False, random_state=None, shrinking=True,
  tol=0.001, verbose=False)
model.fit(X_train_scaled, encoded_y_train)
predictions = model.predict(X_test_scaled)
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(encoded_y_test, predictions))
```

```python
model.coef_
```

```python
pd.Series(abs(model.coef_[0]), index=X_test.columns)
```

```python
top_df = pd.DataFrame({"column":X_test.columns, "importance":model.coef_[0]}).sort_values(by="importance")
top_df
```

```python
#select the top 11 columns for re-training & re-testing
top_11_df = top_df.iloc[-11:,0]
top_11_df 
```

```python
#select the bottom 11 columns for re-training & re-testing
bottom_11_df = top_df.iloc[:11,0]
bottom_11_df 
```

```python
new_top_df = pd.concat([top_11_df, bottom_11_df], ignore_index=True)
new_top_df
```

```python
new_top_df = pd.DataFrame(new_top_df)
new_top_df = new_top_df['column']
```

```python
#convert the names in the column as a list
influential = new_top_df.values.tolist()
influential
```

```python
#get the list to select matching columns in the original df, resulting in only 15 columns of df
df_feature_imp = df[influential]
df_feature_imp
```

```python
# Set the new X
X_fi = df_feature_imp
```

```python
#Create a train test split
X_fi_train, X_fi_test, y_train, y_test = train_test_split(
    X_fi, y, random_state=1)
X_fi_train.head()
```

```python
#Scale the new X data
X_fi_scaler = MinMaxScaler().fit(X_fi_train)
X_fi_train_scaled = X_fi_scaler.transform(X_fi_train)
X_fi_test_scaled = X_fi_scaler.transform(X_fi_test)
```

```python
#train the model
model.fit(X_fi_train_scaled, encoded_y_train)
predictions = model.predict(X_fi_test_scaled)
```

```python
print(classification_report(encoded_y_test, predictions))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001],'kernel': ['rbf', 'linear']}
grid = GridSearchCV(model, param_grid, refit = True, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_fi_train_scaled, encoded_y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
best_model=grid.best_estimator_
best_model
```
