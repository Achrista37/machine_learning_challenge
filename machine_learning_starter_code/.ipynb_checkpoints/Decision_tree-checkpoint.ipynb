```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
#set all columns to be displayed
pd.set_option('display.max_columns', None)
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
X = df.drop("koi_disposition", axis=1)
y = df["koi_disposition"]
print(X.shape, y.shape)
```

```python
df.koi_disposition.unique()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
X_train.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
len(X_train_scaled[0])
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
```

```python
label_encoder.fit(y_test)
encoded_y_test = label_encoder.transform(y_test)
```

```python
 from tensorflow.keras.utils import to_categorical
```

```python
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
y_train_categorical
```

# Train the Model



```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train_scaled, encoded_y_train)
clf.score(X_train_scaled, encoded_y_train)
```

```python
clf.fit(X_train_scaled, encoded_y_train)
```

```python
print(f"Training Data Score: {clf.score(X_train_scaled, encoded_y_train)}")
print(f"Testing Data Score: {clf.score(X_test_scaled, encoded_y_test)}")
```

```python
clf.feature_importances_
```

```python
top_df = pd.DataFrame({"column":X_test.columns, "importance":clf.feature_importances_}).sort_values(by="importance")
```

```python
#select the top 15 columns for re-training & re-testing
top_22_df = top_df.iloc[-22:]
top_22_df = top_22_df['column']
```

```python
#convert the names in the column as a list
top_22 = top_22_df.values.tolist()
top_22
```

```python
#get the list to select matching columns in the original df, resulting in only 15 columns of df
df_feature_imp = df[top_22]
df_feature_imp
```

```python
# Set the new X
X_fi = df_feature_imp
```

```python
#Create a train test split
X_fi_train, X_fi_test, y_train, y_test = train_test_split(
    X_fi, y, random_state=1)
X_fi_train.head()
```

```python
#Scale the new X data
X_fi_scaler = MinMaxScaler().fit(X_fi_train)
X_fi_train_scaled = X_fi_scaler.transform(X_fi_train)
X_fi_test_scaled = X_fi_scaler.transform(X_fi_test)
```

```python
#train the model
clf_fi = tree.DecisionTreeClassifier()
clf_fi = clf_fi.fit(X_fi_train_scaled, encoded_y_train)
clf_fi.score(X_fi_test_scaled, encoded_y_test)
```
